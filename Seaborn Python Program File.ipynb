```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Data = pd.read_csv('https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv')
```

```python
Data.head()
```

```python
New_data = pd.pivot_table(Data,values='lifeExp', index='continent', columns='year')
New_data.columns.name = ' ' 
New_data
```

```python
New_data = pd.DataFrame(New_data.astype(int))
New_data
```

```python
New_data.info()
```

```python
plt.figure(figsize=(18,8))
sns.heatmap(data=New_data,annot=True)
plt.title('HeatMap showing Life expectancy of continents per year')
```
